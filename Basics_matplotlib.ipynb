```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### A first and simple plotting approach

```python
# data
x = np.arange(0, 2*np.pi, 0.1)
y = np.sin(x)

# STEP 1: create a figure object
fig = plt.figure()

# STEP 2: add data
# Plot x and y, label the line with 'sin(x)'
plt.plot(x, y, label='sin(x)')

# STEP 3: add structural elements
plt.legend()

# STEP 4: display/save as file
plt.show()
plt.savefig('my_very_first_plot.pdf')
```

```python
# Sample data
# Generate an array of 100 equally spaced values between 0 and 2*pi
# The resulting array represents the x-values for a full cycle (0 to 2*pi)
x = np.linspace(0, 2*np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create a figure with 1 row and 2 columns of subplots
fig, ax = plt.subplots(1, 3, figsize=(10, 4), dpi=100)

# Plot on the first subplot
# Plot sin(x) on the first subplot, assign label for legend
ax[0].plot(x, y1, label='sin(x)', color='blue')

# Set the title for the first subplot
ax[0].set_title('Sin Subplot')

# Set the label for the X-axis
ax[0].set_xlabel('X-axis')

# Set the label for the Y-axis
ax[0].set_ylabel('Y-axis')

# Add a legend to the first subplot using labels specified in the plot command
ax[0].legend()

# Display grid
ax[0].grid(True)

# Plot on the second subplot
ax[1].plot(x, y2, label='cos(x)', color='red')
ax[1].set_title('Cos Subplot')
ax[1].set_xlabel('X-axis')
ax[1].set_ylabel('Y-axis')
ax[1].legend()
# ax[1].grid(True)  # Compared with grid

# Plot on the third subplot with log scaling on both axes
ax[2].loglog(x, y1, label='sin(x)', color='yellow')
ax[2].set_title('Loglog Scale')
ax[2].set_xlabel('X-axis (log scale)')
ax[2].set_ylabel('Y-axis (log scale)')
ax[2].legend()
ax[2].grid(True)  # Display grid

# Overall figure settings
fig.suptitle('Example Figure with Different Axis Scaling')

# Adjust layout to prevent overlap
fig.tight_layout()

# Show the figure
plt.show()
```

### More relevant structrual elements

```python
# Create a figure with 1 row and 3 columns of subplots
fig, ax = plt.subplots(1, 3, figsize=(18, 4))

# Plot on the first subplot
ax[0].plot(x, y1, label='sin(x)', color='blue')
ax[0].set_title('Linear Scale')
ax[0].set_xlabel('X-axis')
ax[0].set_ylabel('Y-axis')
ax[0].legend()
ax[0].grid(True)

# Set the X-axis limits for the first subplot
ax[0].set_xlim(0, 2*np.pi)

# Set the Y-axis limits for the first subplot
ax[0].set_ylim(-1, 1)

# Plot on the second subplot with log scaling on the x-axis
ax[1].semilogx(x, y1, label='sin(x)', color='green')
ax[1].set_title('Semilogx Scale')
ax[1].set_xlabel('X-axis (log scale)')
ax[1].set_ylabel('Y-axis')
ax[1].legend()
ax[1].grid(True)  # Display grid
ax[1].set_xlim(0.1, 2*np.pi)
ax[1].set_ylim(-1, 1)

# Plot on the third subplot with log scaling on both axes
ax[2].loglog(x, y1, label='sin(x)', color='red')
ax[2].set_title('Loglog Scale')
ax[2].set_xlabel('X-axis (log scale)')
ax[2].set_ylabel('Y-axis (log scale)')
ax[2].legend()
ax[2].grid(True)  # Display grid
ax[2].set_xlim(0.1, 2*np.pi)
ax[2].set_ylim(0.1, 1)

# Overall figure settings
fig.suptitle('Example Figure with Different Axis Scaling')
fig.tight_layout()

# Display the figure
plt.show()
```

## Different types of plots

```python
# Generate some sample data
x = np.linspace(0, 2*np.pi, 100)
y_sin = np.sin(x)
y_cos = np.cos(x)
y_exp = np.exp(x)
y_log = np.log(x + 1e-10)  # Adding a small constant to avoid log(0) error

# Create a figure and axis
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

# Line plot
axes[0, 0].plot(x, y_sin, label='sin(x)', color='blue', linestyle='-', linewidth=2, marker='o', markersize=5, alpha=0.8)
axes[0, 0].plot(x, y_cos, label='cos(x)', color='red', linestyle='--', linewidth=2, marker='x', markersize=5, alpha=0.8)
axes[0, 0].set_title('Line Plot')
axes[0, 0].set_xlabel('X-axis')
axes[0, 0].set_ylabel('Y-axis')
axes[0, 0].legend()

# Scatter plot
axes[0, 1].scatter(x, y_exp, label='exp(x)', color='green', marker='o')
axes[0, 1].set_title('Scatter Plot')
axes[0, 1].set_xlabel('X-axis')
axes[0, 1].set_ylabel('Y-axis')
axes[0, 1].legend()

# Bar plot
categories = ['Category A', 'Category B', 'Category C']
values = [5, 12, 8]
axes[1, 0].bar(categories, values, color=['orange', 'purple', 'pink'])
axes[1, 0].set_title('Bar Plot')
axes[1, 0].set_xlabel('Categories')
axes[1, 0].set_ylabel('Values')

# Histogram
axes[1, 1].hist(y_log, bins=20, color='brown', edgecolor='black')
axes[1, 1].set_title('Histogram')
axes[1, 1].set_xlabel('X-axis')
axes[1, 1].set_ylabel('Frequency')

# Adjust layout for better spacing
fig.tight_layout()

# Display the plot
plt.show()
```

### Surface plots
- we use `np.meshgrid()` below to transform the `X`- and `Y`-array into an actual 2D grid structure to compute the values to be displayed. Whenever you want to do a surface plot the `X` and `Y` need to be of this 2D structure.

```python
from mpl_toolkits.mplot3d import Axes3D

# Create a meshgrid
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)

# Define a function for the surface
z = np.sin(np.sqrt(x**2 + y**2))

# Create a 3D figure
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
# - cmap: Color map for the surface plot
# - edgecolor: Color of the edges of the surface patches
# - rstride, cstride: Row and column stride for sampling the surface
# - alpha: Opacity of the surface plot (transparency)
surf = ax.plot_surface(x, y, z, cmap='plasma', edgecolor='k', rstride=5, cstride=5, alpha=0.8)

# Add a colorbar which maps values to colors
fig.colorbar(surf, ax=ax, shrink=0.5, aspect=50)

# Set labels
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('3D Surface Plot')

# Show the plot
plt.show()
```

## Customizing `matplotlib`
- `matplotlib` has a large variety of property parameters which are summarized in the so-called `rc` (runtime configuration) settings and determine what your plots look like. They are stored in a global dictionary-like variable `matplotlib.rcParams`. 
- To customize your plots you can dynamically change the default values in there.
- The individual parameters are organized in groups (e.g. `lines`, `axes`, `legend`) and can be accessed via these groups, i.e. parameters are named as `<group>.<parameter>`.
- A full list of the parameters you can change can be displayed with `plt.rcParams`.

```python
# Sample data
x = np.arange(0, 2*np.pi, 0.1 )
y = np.sin( x )

# Modify runtime configuration (rc) settings for lines
plt.rcParams['lines.linewidth'] = 5  # Set linewidth to 5
plt.rcParams['lines.linestyle'] = '--'  # Set linestyle to '--'

# Plotting with modified rcParams
plt.plot(x,y)

plt.show()
```

#### Setting multiple parameters at once
- `matplotlib` also offers the option to set multiple parameters from the same group within a single line.
- In case to reset all of `rcParams` to default values use: `plt.rcParams.update(plt.rcParamsDefault)`.

```python
plt.rc('lines', linewidth=1.5, linestyle='-')
plt.plot(x,y)
plt.show()
```

- Another option is to set up a dictionary containing all the parameter changes that you want to perform. This dictionary is then used to update `rcParams` once.
- The example below sets the font type and size to a $\LaTeX$-style one and allows for $\LaTeX$-math mode expressions in title, labels etc.

```python
params = {'legend.fontsize' : 14,
          'legend.handlelength': 2,
          'text.usetex' : True,
          'text.latex.preamble' : r'\usepackage[official]{eurosym}',    # include more latex packages here if you need them for expressions in title, labels, etc.
          'font.size' : 14,
          'font.family' : 'sans-serif'
         }
plt.rcParams.update(params)

plt.plot(x,y)
plt.show()
```

### setting the color scheme
- When using `plt.plot()` (or any corresponding command) you can directly set the color of the line etc. using the respective keyword. If no color is specified the color is picked from the defined color cycler. This is a list of colors that are used one by one for your plotting elements. As soon as all colors have been used `matplotlib` will start again by selecting the first color in the list.
- If you want all your plots to use the same colors for plotting you can define your own color cycler and use that one for the respective `rc` parameter.
- The used colors can either be predefined `matplotlib`-colors (cf. https://matplotlib.org/stable/gallery/color/named_colors.html) or your very own choices based on RGB- or HEX-representation.

```python
# Sample data
x = np.arange(0, 2*np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

# Example 1: Setting colors directly in plt.plot()
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.plot(x, y_sin, label='sin(x)', color='blue', linestyle='-', linewidth=2)
plt.plot(x, y_cos, label='cos(x)', color='red', linestyle='--', linewidth=2)
plt.title('Setting Colors Directly')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()

# Example 2: Using a custom color cycler
# Define your own color cycler by listing any colors of your choice
custom_cycler = plt.cycler(color=['purple', 'green'])
plt.rcParams['axes.prop_cycle'] = custom_cycler

plt.subplot(1, 2, 2)
plt.plot(x, y_sin, label='sin(x)', linestyle='-', linewidth=2)
plt.plot(x, y_cos, label='cos(x)', linestyle='--', linewidth=2)
plt.title('Using Custom Color Cycler')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()

plt.tight_layout()
plt.show()
```

### Adapting temporary `rc` settings
- In case you want to change some of the properties only for a certain part of your script one can use the `matplotlib.rc_context` context manager.
- In accordance with the context manager, we need the keyword `with` in this case.
- The dictionary of parameter adaptions that is passed is only used for the runtime of the indented block. After that everything goes back to the (default) values we had before.

```python
with plt.rc_context({'lines.linewidth': 7, 'lines.linestyle': ':'}):
    # New parameter settings are valid here
    plt.plot(x,y,label='inside of indented block')

# Back to previous (default) settings
plt.plot(x,0.5*y,label='outside of indented block')
plt.legend()
plt.show()
```
